## Beautifulsoup
- pip install beautifulsoup4
- pip install html5lib

```python
from bs4 import BeautifulSoup
```

```python
html_sample = ' \
<html> \
<body> \
<h1 id="title">Hello World</h1> \
<a href="#" class="link">This is link1</a> \
<a href="# link2" class="link">This is link2</a> \
</body> \
</html>'

soup = BeautifulSoup(html_sample, 'html5lib')
print(soup)
print(type(soup))
print(soup.text)
```

## CSS selectors

```python
# 選取h1 tag，並且把內文print出來
h1 = soup.select('h1')
print(h1)
print(type(h1))
print(h1[0])
print(type(h1[0].text))
print(h1[0].text)
```

```python
# 選取a tag，並且把內文print出來
a = soup.select('a')
print(a)
print(a[0])
print(a[0].text)
```

```python
for a in soup.select('a'):
    print(a.text)
```

```python
# 選取a tag裡面的href
for a in soup.select('a'):
    print(a['href'])
```

```python
# 用tag 的id來選取，ID => #
soup = BeautifulSoup(html_sample, 'html.parser')
print(soup.select('#title'))
```

```python
# 用tag的class來選取，CLASS => .
soup = BeautifulSoup(html_sample, 'html.parser')
print(soup.select('.link'))
```
