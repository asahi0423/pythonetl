```python
import requests as r
from bs4 import BeautifulSoup
import pandas as pd
import datetime as dt
from sqlalchemy import create_engine # database connection
```

```python
res = r.get('http://www.thsrc.com.tw/tw/TimeTable/DailyTimeTable/20160606')
```

```python
soup = BeautifulSoup(res.text, "lxml")
```

```python
table = soup.find('table')
```

```python
table
```

```python
south_bound = pd.read_html(str(table), flavor="lxml")[3]
south_bound.columns = south_bound.iloc[0]
south_bound = south_bound.reindex(south_bound.index.drop(0))
stations = south_bound.columns.get_values()[2:]
for station in stations:
    south_bound[station] = '20160606 ' + south_bound[station].astype(str)
    south_bound[station] = pd.to_datetime(south_bound[station], format='%Y%m%d %H:%M', errors="coerce")
south_bound
#south_bound.ix[:,2]
```

```python
disk_engine = create_engine('sqlite:///hsr_time_table.db') # Initializes database with filename 311_8M.db in current directory
```

```python
south_bound.to_sql('table_201606', disk_engine, if_exists='append')
```
